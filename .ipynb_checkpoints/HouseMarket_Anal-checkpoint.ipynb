```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
PR = np.array([464.7,472.1,494.2,542.9,607.1,665.4,723.5,776.0,843.2,916.2,963.8,1019.0,1085.3,1203.1,1342.5,1450.8,1534.6,1572.7])
L = np.array([416.8,445.4,468.9,514.8,575.6,630.1,683.6,734.3,793.8,861.4,905.9,960.6,1025.1,1138.0,1269.8,1370.1,1444.5,1481.6])
SR = np.array([47.9,26.6,25.3,28.0,31.5,35.3,39.9,41.7,49.4,54.8,57.9,58.5,60.2,65.1,72.7,80.8,90.2,91.1])

IR = np.array([1.25,1.75,1.50,1.25,1.50,2.00,2.50,2.75,3.25,2.50,2.00,3.00,5.00,4.50,3.75,3.25,3.75,4.25])
IR = np.flip(IR)
```

```python
print(len(PR), len(L), len(SR), len(IR))

def VariCalc(data):
    vari = []
    for i in range(len(data)):
        if i == 0:
            vari.append(0)
        else:
            vari.append(data[i]-data[i-1])
    return vari

vari_L = VariCalc(L)
vari_IR = VariCalc(IR)
```

```python
# --Normalize
def MinMaxScaler(data):
	numerator = data - np.min(data,0)
	denominator = np.max(data,0) - np.min(data,0)
	denominator = denominator.astype('float')
	return numerator / denominator
    #return numerator / (denominator+1e-7)

#def MinMaxScaler(data):
#	return data
```

```python
normed_PR = MinMaxScaler(PR)
normed_L = MinMaxScaler(L)
normed_SR = MinMaxScaler(SR)
normed_IR = MinMaxScaler(IR)
normed_vari_L = MinMaxScaler(vari_L)
normed_vari_IR = MinMaxScaler(vari_IR)
```

```python
x = np.arange(2,20,1) + 2000
print(x)
```

```python
plt.rcParams["legend.loc"] = 'upper left'
plt.plot(x, normed_IR,'o-',color='midnightblue',label='Interst rate')
plt.plot(x, normed_L,'o-',color='orange',label='Household loans')
plt.plot(x, normed_vari_IR,'*--',color='blue',alpha=0.5,label='gradient IR')
plt.plot(x, normed_vari_L,'*--',color='red',alpha=0.5,label='gradient HL')
plt.xlabel('Year',fontsize=20)
plt.ylabel('Normalized Value',fontsize=20)

plt.legend(prop={'size':10})
plt.grid(which='major', linestyle='--')
plt.minorticks_on()
```
